## Comments

In Python, simple comments included anywhere in the code start with hash #.

For a longer comment, spanning several lines, we use three single quotes to start the comment and close it with three single quotes. These comments are used to document the code in an "official" way. We will see its usefulness later.

All comments are ignored when compiling the program.

```python
#simple comments
"""
longer comment
sdsdsds
sdsdsdsd

"""
```

## Variables
Variables are memory areas where we will temporarily store our data during the execution of the program. As we have already seen in Python, variables are defined when they are necessary, without the need for a formal definition. Python takes care of determining the data type of the variable at run time.
If we want to know the type of a variable we can use the **type ()** function.

```python
a=2
b=a+3
print (a,b)
```

```python
print (type(a))
print(a)
```

Being dynamically typed, python allows you to assign values of different types to a variable during execution. Therefore, be careful as the type of a variable might not be what you expect. In these cases, we can use the type () function. Depending on the types of variables, operators can perform one function or another. For example, the operator + with numbers and with text.
