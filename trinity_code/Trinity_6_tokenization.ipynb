#TOKENIZATION IN SPACY


```python
# Defining string
text= "I'm working as a Data Scientist in the U.S. and earning $140,000."
```

```python
# Whitespace Tokenizer
print(text.split())
```

```python
import spacy
```

Models in spaCy for English Language as of release 2.3.1:
- **en_core_web_sm:** 11MB
- **en_core_web_md:** 48MB
- **en_core_web_lg:** 746MB

```python
# Loading model
nlp=spacy.load('en_core_web_sm')
```

```python
# creating Doc object
doc=nlp(text)
```

```python
doc
```

```python
type(doc)
```

```python
# spaCy Tokenizer
print([token for token in doc])
```

```python
# Type of token
[type(token) for token in doc]
```

```python
# Tokens as string
print([token.text for token in doc])
```

```python
# Comparing Whitespace and spaCy Tokenizer

# Whitespace Tokenizer
print(text.split())

# spaCy Tokenizer
print([token.text for token in doc])
```

```python
# No. of Tokens in Whitespace Tokenizer
print(len(text.split()))

# No. of Tokens in spaCy's Tokenizer
print(len(doc))
```

# Lemmatization in spaCy

```python
# Defining the string
text="The sky is clear and the stars are twinkling."
```

```python
import spacy
```

```python
# Loading spacy model
nlp=spacy.load('en_core_web_sm')
```

```python
# Creating doc object
doc=nlp(text)
```

```python
# Lemmatizing the text
[(token.text,token.lemma_) for token in doc]
```

```python
text="The moon looks beautiful at night. it's hard to resist its beauty."

# Creating doc object
doc=nlp(text)

# Lemmatizing the text
[(token.text,token.lemma_) for token in doc]
```
